```python
product = []
a = -30
for i in range(0,len(df[df.columns[0]])-1):
    b = a+i
    if b < 0:
        product.append('D'+str(a+i))
    elif b == 0:     
        product.append('D-day')
    else:
        product.append('D+'+str(a+i))
product
```
